 * Absolute Frequency: Number of times each value appears in the data.
 
 Accumulated Frequency: Sum of the absolute frequencies up to the current value.
 
 Relative Frequency: Proportion of the absolute frequency with respect to the total data.
 
 * Relative Frequency (%): Percentage of the relative frequency.
 
 Accumulated Relative Frequency: Sum of the relative frequencies up to the current value.
 
 * Cumulative Relative Frequency (%): Percentage of the accumulated relative frequency with respect to the total.

```python
# Barras Valeria
marcas_clases=[1, 2, 3, 4, 5, 6]
frecuencias=[8, 5, 3, 8, 4, 2]
marcas_texto=["Pumas", "Chivas", "Santos", "America", "Toluca", "Necaxa"]
plt.figure(figsize = (12, 6))
plt.barh(marcas_clases, frecuencias, 
        height = 0.75, edgecolor = "k",
       color = ["#eb445b", "#f5938b", "#f0cdab", "#f1e7c5", "#CD617D", "#b6d4bb"])
plt.yticks(marcas_clases, marcas_texto, fontsize = 15)
plt.xlabel("Frequency", fontsize = 20)
plt.ylabel("Classes", fontsize = 20)
plt.title("Bar Diagram", fontsize = 25)
#plt.grid()
plt.show()
```

```python
datos = [26.67 , 16.67, 10, 26.67, 13.33, 6.66]
marcas_texto = ["Pumas", "Chivas", "Santos", "América", "Toluca", "Necaxa"]
separaciones = [0, 0.2, 0, 0, 0.2, 0]

plt.pie(datos, 
       explode = separaciones, 
       counterclock = False, 
       startangle = 90, 
       autopct = "%0.1f%%", 
       pctdistance = 0.8, 
       colors = ["#eb445b", "#f5938b", "#f0cdab", "#f1e7c5", "#CD617D", "#b6d4bb"], 
       labels = marcas_texto)
plt.show()
```
